```python
import matplotlib.pyplot as plt
from matplotlib import animation
import numpy as np
from IPython.display import HTML
```

```python
fig, ax = plt.subplots(ncols=2)
ax[0].set_xlim(0, 2)
ax[0].set_ylim(-2, 5)
ax[1].set_xlim(0, 100)
ax[1].set_ylim(0, 2)
```

```python
x = np.linspace(0, 2, 91)
t = np.linspace(1, 25, 30)
```

```python
line1 = ax[0].plot([], [])
line2 = ax[1].plot([], [], marker='o')
```

```python
decay = 0.5
```

```python
def animate(i):
    x = np.linspace(0, 3, 1000)
    y = np.sin(1 + 2 * np.pi *(x - i * .01))
    decay = 1/np.log(i + .1)
    line1[0].set_data(x, y)
    line2[0].set_data(i, decay)
    ax[0].fill_between(x, y, -2, color='black')
    ax[0].fill_between(x, y, 5, color='white')
    
```

```python
art = line2[0]
```

```python
anim = animation.FuncAnimation(fig, animate, frames=50)
```

```python
HTML(anim.to_jshtml(fps=30))
```
